```python
# Task 1: Check if a given number is positive or negative
def check_positive_negative(number):
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"

# Task 2: Determine if a person is eligible to vote based on their age
def check_voting_eligibility(age):
    if age >= 18:
        return "Eligible to vote"
    else:
        return "Not eligible to vote"

# Task 3: Find the maximum of two numbers using if-else statements
def find_maximum(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2

# Task 4: Classify a given year as a leap year or not
def check_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return "Leap year"
    else:
        return "Not a leap year"

# Task 5: Check whether a character is a vowel or a consonant
def check_vowel_consonant(char):
    if char.lower() in "aeiou":
        return "Vowel"
    else:
        return "Consonant"

# Task 6: Determine whether a given number is even or odd
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Task 7: Calculate the absolute value of a number without using the `abs()` function
def absolute_value(number):
    if number < 0:
        return -number
    else:
        return number

# Task 8: Determine the largest of three given numbers using if-else statements
def find_largest_of_three(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

# Task 9: Check if a given string is a palindrome
def is_palindrome(string):
    clean_string = ''.join(char.lower() for char in string if char.isalnum())
    return clean_string == clean_string[::-1]

# Task 10: Calculate the grade based on a student's score
def calculate_grade(score):
    if 90 <= score <= 100:
        return "A"
    elif 80 <= score < 90:
        return "B"
    elif 70 <= score < 80:
        return "C"
    elif 60 <= score < 70:
        return "D"
    elif 0 <= score < 60:
        return "F"
    else:
        return "Invalid score"

# Testing the functions
print("Task 1:", check_positive_negative(-5))
print("Task 2:", check_voting_eligibility(20))
print("Task 3:", find_maximum(25, 12))
print("Task 4:", check_leap_year(2024))
print("Task 5:", check_vowel_consonant('a'))
print("Task 6:", check_even_odd(7))
print("Task 7:", absolute_value(-10))
print("Task 8:", find_largest_of_three(14, 27, 18))
print("Task 9:", is_palindrome("A man, a plan, a canal: Panama"))
print("Task 10:", calculate_grade(75))
```

# Nested If-Else Statements:-

```python
# Task 11: Find the largest among three numbers using nested if-else statements
def find_largest_among_three(num1, num2, num3):
    if num1 >= num2:
        if num1 >= num3:
            return num1
        else:
            return num3
    else:
        if num2 >= num3:
            return num2
        else:
            return num3

# Task 12: Determine if a triangle is equilateral, isosceles, or scalene
def classify_triangle(side1, side2, side3):
    if side1 == side2 == side3:
        return "Equilateral"
    elif side1 == side2 or side1 == side3 or side2 == side3:
        return "Isosceles"
    else:
        return "Scalene"

# Task 13: Check if a year is a leap year and also if it is a century year
def check_leap_and_century_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        leap_year_status = "Leap year"
    else:
        leap_year_status = "Not a leap year"

    if year % 100 == 0:
        century_year_status = "Century year"
    else:
        century_year_status = "Not a century year"

    return leap_year_status, century_year_status

# Task 14: Determine if a number is positive, negative, or zero
def check_positive_negative_zero(number):
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"

# Task 15: Check if a person is a teenager (between 13 and 19 years old)
def is_teenager(age):
    if 13 <= age <= 19:
        return "Teenager"
    else:
        return "Not a teenager"

# Task 16: Determine the type of angle based on its measure (acute, obtuse, or right)
def classify_angle(measure):
    if measure < 90:
        return "Acute angle"
    elif measure == 90:
        return "Right angle"
    else:
        return "Obtuse angle"

# Task 17: Calculate the roots of a quadratic equation
def quadratic_roots(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        root1 = (-b + (discriminant)**0.5) / (2*a)
        root2 = (-b - (discriminant)**0.5) / (2*a)
        return "Two distinct real roots:", root1, root2
    elif discriminant == 0:
        root = -b / (2*a)
        return "One real root:", root
    else:
        return "No real roots (complex roots)"

# Task 18: Determine the day of the week based on a user-provided number
def day_of_the_week(number):
    if number == 1:
        return "Monday"
    elif number == 2:
        return "Tuesday"
    elif number == 3:
        return "Wednesday"
    elif number == 4:
        return "Thursday"
    elif number == 5:
        return "Friday"
    elif number == 6:
        return "Saturday"
    elif number == 7:
        return "Sunday"
    else:
        return "Invalid day number"

# Task 19: Check if a year is a leap year and also if it is evenly divisible by 400
def check_leap_and_divisible_by_400(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        leap_year_status = "Leap year"
    else:
        leap_year_status = "Not a leap year"

    if year % 400 == 0:
        divisible_by_400_status = "Divisible by 400"
    else:
        divisible_by_400_status = "Not divisible by 400"

    return leap_year_status, divisible_by_400_status

# Task 20: Check if a given number is prime or not using nested if-else statements
def check_prime_number(number):
    if number > 1:
        for i in range(2, int(number**0.5) + 1):
            if number % i == 0:
                return "Not a prime number"
        return "Prime number"
    else:
        return "Not a prime number"

# Testing the functions
print("Task 11:", find_largest_among_three(15, 8, 20))
print("Task 12:", classify_triangle(5, 5, 5))
print("Task 13:", check_leap_and_century_year(2000))
print("Task 14:", check_positive_negative_zero(-7))
print("Task 15:", is_teenager(16))
print("Task 16:", classify_angle(45))
print("Task 17:", day_of_the_week(3))
print("Task 18:", quadratic_roots(1, -3, 2))
print("Task 19:", check_leap_and_divisible_by_400(2000))
print("Task 20:", check_prime_number(17))
```

```python
# Task 21: Assign grades based on different ranges of scores using elif statements
def assign_grade(score):
    if 90 <= score <= 100:
        return "A"
    elif 80 <= score < 90:
        return "B"
    elif 70 <= score < 80:
        return "C"
    elif 60 <= score < 70:
        return "D"
    elif 0 <= score < 60:
        return "F"
    else:
        return "Invalid score"

# Task 22: Determine the type of a triangle based on its angles
def classify_triangle_by_angles(angle1, angle2, angle3):
    if angle1 + angle2 + angle3 == 180:
        if angle1 == angle2 == angle3:
            return "Equilateral triangle"
        elif angle1 == angle2 or angle1 == angle3 or angle2 == angle3:
            return "Isosceles triangle"
        else:
            return "Scalene triangle"
    else:
        return "Invalid triangle angles"

# Task 23: Categorize a given person's BMI into underweight, normal, overweight, or obese using elif statements
def categorize_bmi(weight, height):
    bmi = weight / (height ** 2)
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    elif bmi >= 30:
        return "Obese"
    else:
        return "Invalid BMI"

# Task 24: Determine whether a given number is positive, negative, or zero using elif statements
def classify_number(number):
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"

# Task 25: Determine the type of a character (uppercase, lowercase, or special) using elif statements
def classify_character(char):
    if char.isupper():
        return "Uppercase"
    elif char.islower():
        return "Lowercase"
    else:
        return "Special character"

# Task 26: Calculate the discounted price based on different purchase amounts using elif statements
def calculate_discounted_price(purchase_amount):
    if purchase_amount < 100:
        discount = 0
    elif 100 <= purchase_amount < 500:
        discount = 0.1
    elif 500 <= purchase_amount < 1000:
        discount = 0.2
    else:
        discount = 0.3

    discounted_price = purchase_amount * (1 - discount)
    return discounted_price

# Task 27: Calculate the electricity bill based on different consumption slabs using elif statements
def calculate_electricity_bill(units_consumed):
    if units_consumed <= 100:
        rate_per_unit = 1.5
    elif 101 <= units_consumed <= 200:
        rate_per_unit = 2.0
    elif 201 <= units_consumed <= 300:
        rate_per_unit = 2.5
    else:
        rate_per_unit = 3.0

    electricity_bill = units_consumed * rate_per_unit
    return electricity_bill

# Task 28: Determine the type of quadrilateral based on its angles and sides using elif statements
def classify_quadrilateral(sides, angles):
    if all(side == sides[0] for side in sides):
        return "Square" if all(angle == 90 for angle in angles) else "Rhombus"
    elif angles[0] == angles[1] == angles[2] == angles[3] == 90:
        return "Rectangle"
    else:
        return "Quadrilateral"

# Task 29: Determine the season based on a user-provided month using elif statements
def determine_season(month):
    if 3 <= month <= 5:
        return "Spring"
    elif 6 <= month <= 8:
        return "Summer"
    elif 9 <= month <= 11:
        return "Autumn"
    elif month == 12 or 1 <= month <= 2:
        return "Winter"
    else:
        return "Invalid month"

# Task 30: Determine the type of a year (leap or common) and month (30 or 31 days) using elif statements
def classify_year_and_month(year, month):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        year_type = "Leap year"
    else:
        year_type = "Common year"

    if month in {4, 6, 9, 11}:
        month_days = 30
    elif month == 2:
        month_days = 29 if year_type == "Leap year" else 28
    else:
        month_days = 31

    return year_type, month_days

# Testing the functions
print("Task 21:", assign_grade(85))
print("Task 22:", classify_triangle_by_angles(60, 60, 60))
print("Task 23:", categorize_bmi(65, 1.75))
print("Task 24:", classify_number(-3))
print("Task 25:", classify_character('@'))
print("Task 26:", calculate_discounted_price(700))
print("Task 27:", calculate_electricity_bill(250))
print("Task 28:", classify_quadrilateral([5, 5, 5, 5], [90, 90, 90, 90]))
print("Task 29:", determine_season(9))
print("Task 30:", classify_year_and_month(2022, 2))

```

```python
import math

# Task 11: Calculate the roots of a quadratic equation
def quadratic_roots(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return "Two distinct real roots:", root1, root2
    elif discriminant == 0:
        root = -b / (2*a)
        return "One real root:", root
    else:
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
        return "Complex roots:", f"{real_part} + {imaginary_part}i", f"{real_part} - {imaginary_part}i"

# Task 12: Determine the day of the week based on the day number (1-7)
def day_of_the_week(day_number):
    days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    if 1 <= day_number <= 7:
        return days[day_number - 1]
    else:
        return "Invalid day number"

# Task 13: Calculate the factorial of a given number using recursion
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

# Task 14: Find the largest among three numbers without using the `max()` function
def find_largest_without_max(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

# Task 15: Simulate a basic ATM transaction menu
def atm_transaction_menu():
    balance = 1000

    while True:
        print("1. Check Balance")
        print("2. Withdraw Money")
        print("3. Deposit Money")
        print("4. Exit")

        choice = int(input("Enter your choice (1-4): "))

        if choice == 1:
            print("Your balance is:", balance)
        elif choice == 2:
            amount = float(input("Enter the amount to withdraw: "))
            if amount > balance:
                print("Insufficient funds!")
            else:
                balance -= amount
                print(f"Withdrawal successful. Remaining balance: {balance}")
        elif choice == 3:
            amount = float(input("Enter the amount to deposit: "))
            balance += amount
            print(f"Deposit successful. New balance: {balance}")
        elif choice == 4:
            print("Exiting ATM. Thank you!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

# Task 16: Check if a given string is a palindrome or not
def is_palindrome(string):
    clean_string = ''.join(char.lower() for char in string if char.isalnum())
    return clean_string == clean_string[::-1]

# Task 17: Calculate the average of a list of numbers, excluding the smallest and largest values
def average_without_extremes(numbers):
    if len(numbers) < 3:
        return "Not enough elements to calculate average."
    else:
        sorted_numbers = sorted(numbers)
        trimmed_numbers = sorted_numbers[1:-1]  # Exclude the smallest and largest values
        return sum(trimmed_numbers) / len(trimmed_numbers)

# Task 18: Convert a given temperature from Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Task 19: Simulate a basic calculator for addition, subtraction, multiplication, and division
def basic_calculator():
    while True:
        print("1. Add")
        print("2. Subtract")
        print("3. Multiply")
        print("4. Divide")
        print("5. Exit")

        choice = int(input("Enter your choice (1-5): "))

        if choice == 5:
            print("Exiting calculator. Thank you!")
            break

        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        if choice == 1:
            result = num1 + num2
        elif choice == 2:
            result = num1 - num2
        elif choice == 3:
            result = num1 * num2
        elif choice == 4:
            if num2 != 0:
                result = num1 / num2
            else:
                print("Cannot divide by zero. Please enter a non-zero second number.")
                continue

        print(f"Result: {result}")

# Task 20: Determine the roots of a cubic equation using the Cardano formula
def cubic_roots(a, b, c, d):
    p = c/a - (b**2)/(3*a**2)
    q = (2*b**3)/(27*a**3) - (b*c)/(3*a**2) + d/a

    delta = (q/2)**2 + (p/3)**3

    if delta > 0:
        u = (-q/2 + math.sqrt(delta))**(1/3)
        v = (-q/2 - math.sqrt(delta))**(1/3)
        root1 = u + v
        return "One real root:", root1
    elif delta == 0:
        u = (-q/2)**(1/3)
        v = (-q/2)**(1/3)
        root1 = u + v
        root2 = -(u + v)/2
        return "Two real roots:", root1, root2
    else:
        theta = math.acos(-math.sqrt((q**2/4) / (-p**3/27)))
        u = 2 * math.sqrt(-p/3) * math.cos(theta/3)
        v = 2 * math.sqrt(-p/3) * math.cos((theta + 2*math.pi)/3)
        w = 2 * math.sqrt(-p/3) * math.cos((theta - 2*math.pi)/3)

        root1 = (u - b/(3*a))
        root2 = (v - b/(3*a))
        root3 = (w - b/(3*a))

        return "Three real roots:", root1, root2, root3

# Testing the functions
print("Task 11:", quadratic_roots(1, -3, 2))
print("Task 12:", day_of_the_week(3))
print("Task 13:", factorial_recursive(5))
print("Task 14:", find_largest_without_max(15, 8, 20))
print("Task 15:")
atm_transaction_menu()
print("Task 16:", is_palindrome("A man, a plan, a canal: Panama"))
print("Task 17:", average_without_extremes([5, 2, 8, 1, 9]))
print("Task 18:", celsius_to_fahrenheit(25))
print("Task 19:")
basic_calculator()
print("Task 20:", cubic_roots(1, -6, 11, -6))
```
