## Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return f"Deposited ${amount}. New balance: ${self.balance}"

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"Withdrew ${amount}. New balance: ${self.balance}"
        else:
            return "Insufficient funds!"

# Menu-driven program
def bank_account_menu():
    account_number = input("Enter account number: ")
    account_holder_name = input("Enter account holder name: ")
    bank_account = BankAccount(account_number, account_holder_name)

    while True:
        print("\nBank Account Menu:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Display Balance")
        print("0. Exit")

        choice = input("Enter your choice (0-3): ")

        if choice == "1":
            amount = float(input("Enter the deposit amount: "))
            print(bank_account.deposit(amount))

        elif choice == "2":
            amount = float(input("Enter the withdrawal amount: "))
            print(bank_account.withdraw(amount))

        elif choice == "3":
            print(f"Current balance: ${bank_account.balance}")

        elif choice == "0":
            print("Exiting the Bank Account Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 3.")

# Run the bank account menu
bank_account_menu()
```

# Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.


```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self):
        # Assume a bonus of 10% of the salary
        bonus_percentage = 0.1
        yearly_bonus = self.salary * bonus_percentage
        return yearly_bonus

    def display_employee_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ${self.salary}")

# Menu-driven program for Employee Management
def employee_management_menu():
    employee_id = input("Enter employee ID: ")
    name = input("Enter employee name: ")
    salary = float(input("Enter employee salary: "))
    employee = Employee(employee_id, name, salary)

    while True:
        print("\nEmployee Management Menu:")
        print("1. Calculate Yearly Bonus")
        print("2. Display Employee Details")
        print("0. Exit")

        choice = input("Enter your choice (0-2): ")

        if choice == "1":
            yearly_bonus = employee.calculate_yearly_bonus()
            print(f"Yearly Bonus: ${yearly_bonus}")

        elif choice == "2":
            employee.display_employee_details()

        elif choice == "0":
            print("Exiting the Employee Management Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 2.")

# Run the employee management menu
employee_management_menu()
```

# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {"Car": 10, "Motorcycle": 5, "Truck": 3}
        self.rented_vehicles = {"Car": 0, "Motorcycle": 0, "Truck": 0}

    def rent_vehicle(self, vehicle_type, quantity):
        if vehicle_type in self.available_vehicles and quantity <= self.available_vehicles[vehicle_type]:
            self.available_vehicles[vehicle_type] -= quantity
            self.rented_vehicles[vehicle_type] += quantity
            return f"Rented {quantity} {vehicle_type}(s)."
        else:
            return f"Sorry, {quantity} {vehicle_type}(s) not available for rent."

    def return_vehicle(self, vehicle_type, quantity):
        if vehicle_type in self.rented_vehicles and quantity <= self.rented_vehicles[vehicle_type]:
            self.available_vehicles[vehicle_type] += quantity
            self.rented_vehicles[vehicle_type] -= quantity
            return f"Returned {quantity} {vehicle_type}(s)."
        else:
            return f"Invalid return request for {quantity} {vehicle_type}(s)."

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle_type, quantity in self.available_vehicles.items():
            print(f"{vehicle_type}: {quantity}")

# Menu-driven program for Vehicle Rental
def vehicle_rental_menu():
    rental_system = VehicleRentalSystem()

    while True:
        print("\nVehicle Rental Menu:")
        print("1. Rent a Vehicle")
        print("2. Return a Vehicle")
        print("3. Display Available Vehicles")
        print("0. Exit")

        choice = input("Enter your choice (0-3): ")

        if choice == "1":
            vehicle_type = input("Enter the vehicle type (Car/Motorcycle/Truck): ")
            quantity = int(input("Enter the quantity to rent: "))
            print(rental_system.rent_vehicle(vehicle_type, quantity))

        elif choice == "2":
            vehicle_type = input("Enter the vehicle type (Car/Motorcycle/Truck): ")
            quantity = int(input("Enter the quantity to return: "))
            print(rental_system.return_vehicle(vehicle_type, quantity))

        elif choice == "3":
            rental_system.display_available_vehicles()

        elif choice == "0":
            print("Exiting the Vehicle Rental Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 3.")

# Run the vehicle rental menu
vehicle_rental_menu()
```

# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def __str__(self):
        return f"{self.title} by {self.author}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        book = Book(title, author)
        self.books.append(book)
        return f"Added '{title}' by {author} to the library."

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.available:
                book.available = False
                return f"Borrowed '{title}' successfully."
        return f"Sorry, '{title}' is not available for borrowing."

    def display_available_books(self):
        available_books = [book for book in self.books if book.available]
        if available_books:
            print("Available Books:")
            for book in available_books:
                print(book)
        else:
            print("No available books in the library.")

# Menu-driven program for Library Catalog
def library_catalog_menu():
    library = Library()

    while True:
        print("\nLibrary Catalog Menu:")
        print("1. Add Book to Library")
        print("2. Borrow Book")
        print("3. Display Available Books")
        print("0. Exit")

        choice = input("Enter your choice (0-3): ")

        if choice == "1":
            title = input("Enter the book title: ")
            author = input("Enter the author's name: ")
            print(library.add_book(title, author))

        elif choice == "2":
            title = input("Enter the book title to borrow: ")
            print(library.borrow_book(title))

        elif choice == "3":
            library.display_available_books()

        elif choice == "0":
            print("Exiting the Library Catalog Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 3.")

# Run the library catalog menu
library_catalog_menu()
```

# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class Product:
    def __init__(self, product_id, name, quantity, price):
        self.product_id = product_id
        self.name = name
        self.quantity = quantity
        self.price = price

    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Quantity: {self.quantity}, Price: ${self.price}"

class InventorySystem:
    def __init__(self):
        self.products = []

    def add_product(self, product_id, name, quantity, price):
        product = Product(product_id, name, quantity, price)
        self.products.append(product)
        return f"Added {name} to the inventory."

    def update_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                return f"Updated quantity for {product.name} to {new_quantity}."
        return f"Product with ID {product_id} not found."

    def display_available_products(self):
        if self.products:
            print("Available Products:")
            for product in self.products:
                print(product)
        else:
            print("No products in the inventory.")

# Menu-driven program for Product Inventory
def product_inventory_menu():
    inventory_system = InventorySystem()

    while True:
        print("\nProduct Inventory Menu:")
        print("1. Add Product to Inventory")
        print("2. Update Product Quantity")
        print("3. Display Available Products")
        print("0. Exit")

        choice = input("Enter your choice (0-3): ")

        if choice == "1":
            product_id = input("Enter the product ID: ")
            name = input("Enter the product name: ")
            quantity = int(input("Enter the product quantity: "))
            price = float(input("Enter the product price: "))
            print(inventory_system.add_product(product_id, name, quantity, price))

        elif choice == "2":
            product_id = input("Enter the product ID to update quantity: ")
            new_quantity = int(input("Enter the new quantity: "))
            print(inventory_system.update_quantity(product_id, new_quantity))

        elif choice == "3":
            inventory_system.display_available_products()

        elif choice == "0":
            print("Exiting the Product Inventory Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 3.")

# Run the product inventory menu
product_inventory_menu()
```

# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        if self.width is None and self.height is None:
            # For 1D shape (e.g., line), area is not applicable
            return "Area calculation not applicable for 1D shape."
        elif self.height is None:
            # For 2D shape (e.g., rectangle), area = length * width
            return f"Area: {self.length * self.width}"
        else:
            # For 3D shape (e.g., cuboid), area = 2 * (length * width + length * height + width * height)
            return f"Surface Area: {2 * (self.length * self.width + self.length * self.height + self.width * self.height)}"

    def calculate_perimeter(self):
        if self.width is None and self.height is None:
            # Perimeter is not applicable for 1D shape
            return "Perimeter calculation not applicable for 1D shape."
        elif self.height is None:
            # For 2D shape (e.g., rectangle), perimeter = 2 * (length + width)
            return f"Perimeter: {2 * (self.length + self.width)}"
        else:
            # Perimeter is not applicable for 3D shape
            return "Perimeter calculation not applicable for 3D shape."

# Menu-driven program for Shape Calculation
def shape_calculation_menu():
    print("Enter the dimensions of the shape:")
    length = float(input("Length: "))
    width = float(input("Width (Enter 0 if not applicable): "))
    height = float(input("Height (Enter 0 if not applicable): "))

    shape = Shape(length, width, height)

    while True:
        print("\nShape Calculation Menu:")
        print("1. Calculate Area")
        print("2. Calculate Perimeter")
        print("0. Exit")

        choice = input("Enter your choice (0-2): ")

        if choice == "1":
            print(shape.calculate_area())

        elif choice == "2":
            print(shape.calculate_perimeter())

        elif choice == "0":
            print("Exiting the Shape Calculation Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 2.")

# Run the shape calculation menu
shape_calculation_menu()
```

#Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average_grade(self):
        if not self.grades:
            return "No grades available for calculation."
        average_grade = sum(self.grades) / len(self.grades)
        return f"Average Grade: {average_grade}"

    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        if self.grades:
            print("Grades:", ", ".join(map(str, self.grades)))
            print(self.calculate_average_grade())
        else:
            print("No grades available.")

# Menu-driven program for Student Management
def student_management_menu():
    student_id = input("Enter student ID: ")
    name = input("Enter student name: ")
    student = Student(student_id, name)

    while True:
        print("\nStudent Management Menu:")
        print("1. Add Grade")
        print("2. Display Student Details")
        print("0. Exit")

        choice = input("Enter your choice (0-2): ")

        if choice == "1":
            grade = float(input("Enter the grade: "))
            student.add_grade(grade)
            print(f"Grade {grade} added successfully.")

        elif choice == "2":
            student.display_student_details()

        elif choice == "0":
            print("Exiting the Student Management Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 2.")

# Run the student management menu
student_management_menu()
```


#Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
class Email:
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.sent = False

    def send_email(self):
        self.sent = True
        return "Email sent successfully."

    def display_email_details(self):
        print("Email Details:")
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print("Status: Sent" if self.sent else "Status: Draft")

# Menu-driven program for Email Management
def email_management_menu():
    sender = input("Enter sender's email address: ")
    recipient = input("Enter recipient's email address: ")
    subject = input("Enter email subject: ")

    email = Email(sender, recipient, subject)

    while True:
        print("\nEmail Management Menu:")
        print("1. Send Email")
        print("2. Display Email Details")
        print("0. Exit")

        choice = input("Enter your choice (0-2): ")

        if choice == "1":
            if not email.sent:
                print(email.send_email())
            else:
                print("Email has already been sent.")

        elif choice == "2":
            email.display_email_details()

        elif choice == "0":
            print("Exiting the Email Management Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 2.")

# Run the email management menu
email_management_menu()
```

# Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, content):
        self.posts.append(content)
        return "Post added successfully."

    def display_posts(self):
        if not self.posts:
            print("No posts available.")
        else:
            print(f"Posts by {self.username}:")
            for i, post in enumerate(self.posts, start=1):
                print(f"{i}. {post}")

    def search_posts_by_keyword(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if matching_posts:
            print(f"Posts containing '{keyword}':")
            for i, post in enumerate(matching_posts, start=1):
                print(f"{i}. {post}")
        else:
            print(f"No posts containing '{keyword}' found.")

# Menu-driven program for Social Media Profile
def social_media_profile_menu():
    username = input("Enter your username: ")
    profile = SocialMediaProfile(username)

    while True:
        print("\nSocial Media Profile Menu:")
        print("1. Add Post")
        print("2. Display Posts")
        print("3. Search Posts by Keyword")
        print("0. Exit")

        choice = input("Enter your choice (0-3): ")

        if choice == "1":
            content = input("Enter your post content: ")
            print(profile.add_post(content))

        elif choice == "2":
            profile.display_posts()

        elif choice == "3":
            keyword = input("Enter the keyword to search for: ")
            profile.search_posts_by_keyword(keyword)

        elif choice == "0":
            print("Exiting the Social Media Profile Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 3.")

# Run the social media profile menu
social_media_profile_menu()
```

# Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date):
        self.tasks.append({"task": task, "due_date": due_date, "completed": False})
        return "Task added successfully."

    def mark_task_completed(self, task_index):
        if 0 <= task_index < len(self.tasks):
            self.tasks[task_index]["completed"] = True
            return f"Task '{self.tasks[task_index]['task']}' marked as completed."
        else:
            return "Invalid task index."

    def display_pending_tasks(self):
        pending_tasks = [task for task in self.tasks if not task["completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for i, task in enumerate(pending_tasks, start=1):
                print(f"{i}. Task: {task['task']}, Due Date: {task['due_date']}")
        else:
            print("No pending tasks.")

# Menu-driven program for ToDo List
def todo_list_menu():
    todo_list = ToDoList()

    while True:
        print("\nToDo List Menu:")
        print("1. Add Task")
        print("2. Mark Task as Completed")
        print("3. Display Pending Tasks")
        print("0. Exit")

        choice = input("Enter your choice (0-3): ")

        if choice == "1":
            task = input("Enter the task: ")
            due_date = input("Enter the due date: ")
            print(todo_list.add_task(task, due_date))

        elif choice == "2":
            task_index = int(input("Enter the index of the task to mark as completed: "))
            print(todo_list.mark_task_completed(task_index - 1))

        elif choice == "3":
            todo_list.display_pending_tasks()

        elif choice == "0":
            print("Exiting the ToDo List Menu. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 0 and 3.")

# Run the ToDo list menu
todo_list_menu()
```
