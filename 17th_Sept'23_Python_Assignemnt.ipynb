# 7 Write a Python program that calculates the factorial of a number using a for loop

```python
# Function to calculate factorial
def calculate_factorial(number):
    factorial = 1

    # Check if the number is negative, zero, or positive
    if number < 0:
        return "Factorial is not defined for negative numbers"
    elif number == 0:
        return 1
    else:
        # Calculate factorial using a for loop
        for i in range(1, number + 1):
            factorial *= i

        return factorial

# Input number for which factorial needs to be calculated
input_number = int(input("Enter a number to calculate its factorial: "))

# Calculate factorial and display the result
result = calculate_factorial(input_number)
print(f"The factorial of {input_number} is: {result}")
```

```python
# Task 1: Iterate through the characters of a string using a for loop
def iterate_string(input_string):
    print("Iterating through characters of the string:")
    for char in input_string:
        print(char, end=' ')
    print("\n")


# Task 2: Find the largest number in a list using a for loop
def find_largest_number(num_list):
    largest_number = float('-inf')
    for num in num_list:
        if num > largest_number:
            largest_number = num
    print("The largest number in the list is:", largest_number, "\n")


# Task 3: Print the Fibonacci sequence up to a specified limit using a for loop
def fibonacci_sequence(limit):
    a, b = 0, 1
    fib_sequence = []
    while a <= limit:
        fib_sequence.append(a)
        a, b = b, a + b
    print("Fibonacci sequence up to", limit, ":", fib_sequence, "\n")


# Task 4: Count the number of vowels in a given string using a for loop
def count_vowels(input_string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in input_string:
        if char in vowels:
            count += 1
    print("Number of vowels in the string:", count, "\n")


# Task 5: Generate a multiplication table for a given number using a for loop
def multiplication_table(num):
    print(f"Multiplication table for {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num*i}")
    print("\n")


# Task 6: Reverse a list using a for loop
def reverse_list(input_list):
    reversed_list = []
    for item in reversed(input_list):
        reversed_list.append(item)
    print("Reversed list:", reversed_list, "\n")


# Task 7: Find the common elements between two lists using a for loop
def common_elements(list1, list2):
    common_elements_list = [item for item in list1 if item in list2]
    print("Common elements between the two lists:", common_elements_list, "\n")


# Task 8: Explain how to use a for loop to iterate through the keys and values of a dictionary in Python
def iterate_dict(sample_dict):
    print("Iterating through keys and values of the dictionary:")
    for key, value in sample_dict.items():
        print(f"Key: {key}, Value: {value}")
    print("\n")


# Task 9: Find the GCD (Greatest Common Divisor) of two numbers using a for loop
def find_gcd(num1, num2):
    while num2:
        num1, num2 = num2, num1 % num2
    return num1


# Task 10: Check if a string is a palindrome using a for loop
def is_palindrome(input_str):
    reversed_str = input_str[::-1]
    return input_str == reversed_str


# Task 11: Remove duplicates from a list using a for loop
def remove_duplicates(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list


# Task 12: Count the number of words in a sentence using a for loop
def count_words(sentence):
    word_count = len(sentence.split())
    return word_count


# Task 13: Find the sum of all odd numbers from 1 to 50 using a for loop
def sum_odd_numbers():
    odd_sum = 0
    for num in range(1, 51, 2):
        odd_sum += num
    print("Sum of all odd numbers from 1 to 50:", odd_sum, "\n")


# Task 14: Check if a given year is a leap year using a for loop
def is_leap_year(year):
    for _ in range(4):  # Checking divisibility by 4 for 4 consecutive years
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        year += 1
    return False


# Task 15: Calculate the square root of a number using a for loop
def calculate_square_root(num):
    if num >= 0:
        sqrt = num ** 0.5
        print(f"The square root of {num} is: {sqrt}\n")
    else:
        print("Cannot calculate the square root of a negative number.\n")


# Task 16: Find the LCM (Least Common Multiple) of two numbers using a for loop
def find_lcm(num1, num2):
    def calculate_gcd(x, y):
        while y:
            x, y = y, x % y
        return x

    gcd = calculate_gcd(num1, num2)
    lcm = abs(num1 * num2) // gcd
    return lcm


# Examples of using the functions:
input_string = "Hello, World!"
num_list = [23, 45, 67, 12, 89, 34, 56]
limit_fibonacci = 100
sample_dict = {"a": 1, "b": 2, "c": 3}
num1_gcd = 36
num2_gcd = 48
input_str_palindrome = "radar"
input_list_duplicates = [1, 2, 3, 2, 4, 5, 6, 4]
sentence_word_count = "This is a sample sentence."
leap_year_check = 2024
sqrt_number = 25
num1_lcm = 12
num2_lcm = 18

# Task 1
iterate_string(input_string)

# Task 2
find_largest_number(num_list)

# Task 3
fibonacci_sequence(limit_fibonacci)

# Task 4
count_vowels(input_string)

# Task 5
multiplication_table(7)

# Task 6
reverse_list(num_list)

# Task 7
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
common_elements(list1, list2)

# Task 8
iterate_dict(sample_dict)

# Task 9
gcd_result = find_gcd(num1_gcd, num2_gcd)
print(f"The GCD of {num1_gcd} and {num2_gcd} is: {gcd_result}\n")

# Task 10
if is_palindrome(input_str_palindrome):
    print(f"{input_str_palindrome} is a palindrome.\n")
else:
    print(f"{input_str_palindrome} is not a palindrome.\n")

# Task 11
unique_list = remove_duplicates(input_list_duplicates)
print(f"List after removing duplicates: {unique_list}\n")

# Task 12
word_count_result = count_words(sentence_word_count)
print(f"Number of words in the sentence: {word_count_result}\n")

# Task 13
sum_odd_numbers()

# Task 14
if is_leap_year(leap_year_check):
    print(f"{leap_year_check} is a leap year.\n")
else:
    print(f"{leap_year_check} is not a leap year.\n")

# Task 15
calculate_square_root(sqrt_number)

# Task 16
lcm_result = find_lcm(num1_lcm, num2_lcm)
print(f"The LCM of {num1_lcm} and {num2_lcm} is: {lcm_result}\n")
```

```python
def square_elements():
    numbers = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]

    # Use map() to square each element of the list
    squared_numbers = map(lambda x: x**2, numbers)

    # Convert the map object to a list (for Python 3)
    squared_numbers_list = list(squared_numbers)

    # Print the result
    print("Squared numbers:", squared_numbers_list)


def uppercase_names():
    names = input("Enter a list of names separated by spaces: ").split()

    # Use map() to convert each name to uppercase
    uppercase_names = map(str.upper, names)

    # Convert the map object to a list (for Python 3)
    uppercase_names_list = list(uppercase_names)

    # Print the result
    print("Uppercase names:", uppercase_names_list)


def word_lengths():
    words = input("Enter a list of words separated by spaces: ").split()

    # Use map() to calculate the length of each word
    word_lengths = map(len, words)

    # Convert the map object to a list (for Python 3)
    word_lengths_list = list(word_lengths)

    # Print the result
    print("Word lengths:", word_lengths_list)


def custom_function_on_multiple_lists():
    list1 = [int(x) for x in input("Enter the first list of numbers separated by spaces: ").split()]
    list2 = [int(x) for x in input("Enter the second list of numbers separated by spaces: ").split()]

    # Use map() to apply a custom function to corresponding elements of the lists
    result = map(lambda x, y: x + y, list1, list2)

    # Convert the map object to a list (for Python 3)
    result_list = list(result)

    # Print the result
    print("Result of custom function on multiple lists:", result_list)


def celsius_to_fahrenheit():
    temperatures_celsius = [float(x) for x in input("Enter a list of temperatures in Celsius separated by spaces: ").split()]

    # Use map() to convert temperatures from Celsius to Fahrenheit
    temperatures_fahrenheit = map(lambda x: (x * 9/5) + 32, temperatures_celsius)

    # Convert the map object to a list (for Python 3)
    temperatures_fahrenheit_list = list(temperatures_fahrenheit)

    # Print the result
    print("Temperatures in Fahrenheit:", temperatures_fahrenheit_list)


def round_float_numbers():
    float_numbers = [float(x) for x in input("Enter a list of floating-point numbers separated by spaces: ").split()]

    # Use map() to round each element to the nearest integer
    rounded_numbers = map(round, float_numbers)

    # Convert the map object to a list (for Python 3)
    rounded_numbers_list = list(rounded_numbers)

    # Print the result
    print("Rounded numbers:", rounded_numbers_list)


# Menu-driven program
while True:
    print("\nMenu:")
    print("1. Explain the purpose of the `map()` function in Python and provide an example.")
    print("2. Write a Python program that uses the `map()` function to square each element of a list of numbers.")
    print("3. How does the `map()` function differ from a list comprehension in Python, and when would you choose one over the other?")
    print("4. Create a Python program that uses the `map()` function to convert a list of names to uppercase.")
    print("5. Write a Python program that uses the `map()` function to calculate the length of each word in a list of strings.")
    print("6. How can you use the `map()` function to apply a custom function to elements of multiple lists simultaneously in Python?")
    print("7. Create a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.")
    print("8. Write a Python program that uses the `map()` function to round each element of a list of floating-point numbers to the nearest integer.")
    print("0. Exit")

    choice = input("Enter your choice (0-8): ")

    if choice == "1":
        # Explanation of the `map()` function is provided in the initial comment.
        print("The `map()` function in Python is used to apply a specified function to all items in an iterable and returns an iterable map object.")
        print("Here's an example:")
        # Example code for question 1
        numbers = [1, 2, 3, 4, 5]
        squared_numbers = map(lambda x: x**2, numbers)
        squared_numbers_list = list(squared_numbers)
        print("Squared numbers:", squared_numbers_list)

    elif choice == "2":
        square_elements()
    elif choice == "3":
        # Explanation for question 3
        print("The `map()` function and list comprehension are both used to transform elements in an iterable.")
        print("However, list comprehensions are often considered more readable and concise for simple transformations.")
        print("You might choose `map()` when applying a specific function to all elements in an iterable, especially when working with existing functions.")
    elif choice == "4":
        uppercase_names()
    elif choice == "5":
        word_lengths()
    elif choice == "6":
        custom_function_on_multiple_lists()
    elif choice == "7":
        celsius_to_fahrenheit()
    elif choice == "8":
        round_float_numbers()
    elif choice == "0":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 0 and 8.")
```

# Reduce

```python
from functools import reduce

def product_of_elements():
    numbers = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]

    # Use reduce() to find the product of all elements in the list
    product = reduce(lambda x, y: x * y, numbers, 1)

    # Print the result
    print("Product of elements:", product)


def max_element():
    numbers = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]

    # Use reduce() to find the maximum element in the list
    max_value = reduce(lambda x, y: x if x > y else y, numbers)

    # Print the result
    print("Maximum element:", max_value)


def concatenate_strings():
    strings = input("Enter a list of strings separated by spaces: ").split()

    # Use reduce() to concatenate strings into a single string
    concatenated_string = reduce(lambda x, y: x + y, strings, "")

    # Print the result
    print("Concatenated string:", concatenated_string)


def factorial():
    n = int(input("Enter a number to calculate its factorial: "))

    # Use reduce() to calculate the factorial of a number
    factorial_result = reduce(lambda x, y: x * y, range(1, n + 1), 1)

    # Print the result
    print(f"Factorial of {n}: {factorial_result}")


def gcd_of_numbers():
    numbers = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]

    # Function to calculate GCD (Greatest Common Divisor) of two numbers
    def gcd(x, y):
        while y:
            x, y = y, x % y
        return x

    # Use reduce() to find the GCD of a list of numbers
    gcd_result = reduce(gcd, numbers)

    # Print the result
    print("GCD of numbers:", gcd_result)


def sum_of_digits():
    number = int(input("Enter a number to find the sum of its digits: "))

    # Use reduce() to find the sum of the digits of a number
    sum_of_digits_result = reduce(lambda x, y: int(x) + int(y), str(number), 0)

    # Print the result
    print(f"Sum of digits of {number}: {sum_of_digits_result}")


# Menu-driven program
while True:
    print("\nMenu:")
    print("1. What is the `reduce()` function in Python, and what module should you import to use it? Provide an example of its basic usage.")
    print("2. Write a Python program that uses the `reduce()` function to find the product of all elements in a list.")
    print("3. Create a Python program that uses `reduce()` to find the maximum element in a list of numbers.")
    print("4. How can you use the `reduce()` function to concatenate a list of strings into a single string?")
    print("5. Write a Python program that calculates the factorial of a number using the `reduce()` function.")
    print("6. Create a Python program that uses `reduce()` to find the GCD (Greatest Common Divisor) of a list of numbers.")
    print("7. Write a Python program that uses the `reduce()` function to find the sum of the digits of a given number.")
    print("0. Exit")

    choice = input("Enter your choice (0-7): ")

    if choice == "1":
        # Explanation of the `reduce()` function is provided in the initial comment.
        print("The `reduce()` function in Python is used to apply a specified function to the items of an iterable from left to right.")
        print("It reduces the iterable to a single accumulated result.")
        print("Example of basic usage:")
        # Example code for question 1
        from functools import reduce
        numbers = [1, 2, 3, 4, 5]
        product = reduce(lambda x, y: x * y, numbers)
        print("Product of elements:", product)

    elif choice == "2":
        product_of_elements()
    elif choice == "3":
        max_element()
    elif choice == "4":
        concatenate_strings()
    elif choice == "5":
        factorial()
    elif choice == "6":
        gcd_of_numbers()
    elif choice == "7":
        sum_of_digits()
    elif choice == "0":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 0 and 7.")
```

# Recursion

```python
def factorial_recursive(n):
    # Base case: factorial of 0 is 1
    if n == 0:
        return 1
    # Recursive case: n! = n * (n-1)!
    else:
        return n * factorial_recursive(n - 1)


def fibonacci_recursive(n):
    # Base cases: Fibonacci of 0 is 0, and Fibonacci of 1 is 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # Recursive case: F(n) = F(n-1) + F(n-2)
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)


def sum_of_elements_recursive(lst):
    # Base case: sum of an empty list is 0
    if not lst:
        return 0
    # Recursive case: sum = first element + sum of the rest of the elements
    else:
        return lst[0] + sum_of_elements_recursive(lst[1:])


def gcd_recursive(a, b):
    # Base case: GCD(a, 0) = a
    if b == 0:
        return a
    # Recursive case: GCD(a, b) = GCD(b, a % b)
    else:
        return gcd_recursive(b, a % b)


def reverse_string_recursive(s):
    # Base case: an empty string is its own reverse
    if not s:
        return s
    # Recursive case: reverse the rest of the string + first character
    else:
        return reverse_string_recursive(s[1:]) + s[0]


def power_recursive(x, n):
    # Base case: x^0 = 1
    if n == 0:
        return 1
    # Recursive case: x^n = x * x^(n-1)
    else:
        return x * power_recursive(x, n - 1)


def permutations_recursive(s):
    # Base case: permutations of an empty string is an empty list
    if not s:
        return ['']
    else:
        result = []
        # Iterate over each character in the string
        for i in range(len(s)):
            # Fix the current character and find permutations of the rest
            rest_permutations = permutations_recursive(s[:i] + s[i+1:])
            # Add the fixed character to the front of each permutation
            result += [s[i] + perm for perm in rest_permutations]
        return result


def is_palindrome_recursive(s):
    # Base case: a string with 0 or 1 character is a palindrome
    if len(s) <= 1:
        return True
    # Recursive case: check if the first and last characters are the same
    else:
        return s[0] == s[-1] and is_palindrome_recursive(s[1:-1])


def combinations_recursive(elements, k):
    # Base case: combinations of an empty list or k = 0 is an empty list
    if k == 0 or not elements:
        return ['']
    else:
        result = []
        # Include the first element in the combination and find combinations of the rest
        with_first_element = [elements[0] + combo for combo in combinations_recursive(elements[1:], k - 1)]
        # Exclude the first element and find combinations of the rest
        without_first_element = combinations_recursive(elements[1:], k)
        result.extend(with_first_element)
        result.extend(without_first_element)
        return result


# Menu-driven program
while True:
    print("\nMenu:")
    print("1. Explain the concept of recursion in Python. How does it differ from iteration?")
    print("2. Write a Python program to calculate the factorial of a number using recursion.")
    print("3. Create a recursive Python function to find the nth Fibonacci number.")
    print("4. Write a recursive Python function to calculate the sum of all elements in a list.")
    print("5. How can you prevent a recursive function from running indefinitely, causing a stack overflow error?")
    print("6. Create a recursive Python function to find the greatest common divisor (GCD) of two numbers using the Euclidean algorithm.")
    print("7. Write a recursive Python function to reverse a string.")
    print("8. Create a recursive Python function to calculate the power of a number (x^n).")
    print("9. Write a recursive Python function to find all permutations of a given string.")
    print("10. Write a recursive Python function to check if a string is a palindrome.")
    print("11. Create a recursive Python function to generate all possible combinations of a list of elements.")
    print("0. Exit")

    choice = input("Enter your choice (0-11): ")

    if choice == "1":
        print("Recursion is a programming concept where a function calls itself in order to solve a problem.")
        print("It differs from iteration (looping) in that it uses a function's own calls to repeat a set of instructions.")
        print("Recursion is often used when a problem can be broken down into smaller, similar sub-problems.")

    elif choice == "2":
        n = int(input("Enter a number to calculate its factorial: "))
        print("Factorial:", factorial_recursive(n))

    elif choice == "3":
        n = int(input("Enter the value of n to find the nth Fibonacci number: "))
        print(f"The {n}th Fibonacci number is:", fibonacci_recursive(n))

    elif choice == "4":
        lst = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]
        print("Sum of elements:", sum_of_elements_recursive(lst))

    elif choice == "5":
        print("To prevent a recursive function from running indefinitely, you can ensure that there is a base case.")
        print("The base case is a condition that, when met, stops the recursion and provides a result.")
        print("Without a base case, the function will keep calling itself, leading to a stack overflow error.")

    elif choice == "6":
        a = int(input("Enter the first number: "))
        b = int(input("Enter the second number: "))
        print("GCD of", a, "and", b, "is:", gcd_recursive(a, b))

    elif choice == "7":
        s = input("Enter a string to reverse: ")
        print("Reversed string:", reverse_string_recursive(s))

    elif choice == "8":
        x = float(input("Enter the base (x): "))
        n = int(input("Enter the exponent (n): "))
        print(f"{x}^{n} is:", power_recursive(x, n))

    elif choice == "9":
        string_permutations = permutations_recursive(input("Enter a string: "))
        print("All permutations of the string:", string_permutations)

    elif choice == "10":
        s = input("Enter a string to check if it's a palindrome: ")
        if is_palindrome_recursive(s):
            print("The string is a palindrome.")
        else:
            print("The string is not a palindrome.")

    elif choice == "11":
        elements = input("Enter a list of elements separated by spaces: ").split()
        k = int(input("Enter the length of combinations (k): "))
        combinations = combinations_recursive(elements, k)
        print(f"All combinations of length {k}:", combinations)

    elif choice == "0":
        print("Exiting the program. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number between 0 and 11.")
```
